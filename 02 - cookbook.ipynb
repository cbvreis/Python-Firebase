```python
def make_fake_data(n_rows: int) -> pd.DataFrame:
        faker = Faker('pt_BR')
        
        aux = faker.profile()
        aux.pop('address')
        aux.pop('mail')
        aux.pop('residence')
        cols = list(aux.keys())
        
        array = list()
        for _ in range(n_rows):
            aux = faker.profile()
            aux.pop('address')
            aux.pop('mail')
            aux.pop('residence')
            
            array.append(list(aux.values()))
        
        res = pd.DataFrame(columns=cols, data=array)
        
        return res
```

```python
%%timeit
df = make_fake_data(50000)
```
